# Histograms

Plotting histograms with matplotlib and seaborn

https://matplotlib.org/3.2.2/api/_as_gen/matplotlib.pyplot.hist.html

https://seaborn.pydata.org/generated/seaborn.distplot.html#seaborn.distplot

```python
import matplotlib.pyplot as plt
import seaborn as sns
import numpy as np
```

## matplotlib
Simplest example with matplotlib \
Creating two arrays: x1 is array of random ints [0,10) drawn from uniform distribution; x2 is 100 floats drawn from a normal distribution (mean=0, std=1)

```python
x1 = np.random.randint(0, 10, 100)
x2 = np.random.randn(100)
print(x1)
print(x2)
```

```python
plt.hist(x1)
plt.show()
plt.hist(x2)
plt.show()
```

Number of bins is determined automatically, but can be set explicitly; matplotlib attributes (title, x_label, etc.) can be set normally

```python
plt.hist(x2, bins=20)
plt.title('Normal distribution')
plt.show()
```

When there are only a few different categories to display counts for, it is often clearer to display the exact counts for each category. Problematically the x_ticks are not aligned, because the "bins" parameter specifies the number of bins for values between values of 0-9 (more on this here: https://stackoverflow.com/questions/27083051/matplotlib-xticks-not-lining-up-with-histogram). 

```python
plt.grid(True)
plt.hist(x1, bins=10)
plt.xticks(range(10))
plt.show()
```

Adding one extra bin effectively makes it a left aligned histogram

```python
plt.grid(True)
plt.hist(x1, bins=range(11))
plt.xticks(range(10))
plt.show()
```

To center align, shift bins to the left by 0.5 (and keep the extra bin)

```python
plt.grid(True)
bins = np.arange(11) - 0.5
plt.hist(x1, bins)
plt.xticks(range(10))
plt.show()
```

## seaborn
Seaborn distplot is based on matplotlib's hist plot, so it shares many features. Calculates relative counts instead of absolute counts

Like many seaborn figures, it calculates kernel density and adds this automatically

```python
sns.distplot(x1)
plt.show()
sns.distplot(x2)
plt.show()
```

Can use the nice color palettes of seaborn; same bin cetering trick works as well

```python
sns.set(style="darkgrid")
bins = np.arange(11) - 0.5
sns.distplot(x1, bins)
plt.xticks(range(10))
plt.show()
```
