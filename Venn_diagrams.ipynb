# Venn diagrams

Plotting Venn diagrams using matplotlib_venn (not part of the standard matplotlib library, needs to be installed: https://pypi.org/project/matplotlib-venn/)

```python
import matplotlib.pyplot as plt
from matplotlib_venn import venn2, venn2_circles
```

## Venn diagrams with 2 groups

This is the most basic example, where each value is explicitely defined. The subsets are (count_of_left_circle_exclusively, count_of_right_circle_exclusively, count_of_left-right_circle_intersections)

```python
venn2(subsets=(10, 5, 2))
plt.show()
```

The counts can also be inferred from sets given to the venn2 method. The arguments given have to be specifically of class 'set' - an error is returned if you pass a list.

```python
films_i_like = set(['breakfast_at_tiffanys', 'road_house'])
films_megan_likes = set(['inception', 'breakfast_at_tiffanys', 'thor_ragnarok'])

venn2([films_i_like, films_megan_likes])
plt.show()
```

To zhuzh it up add title, group names, change colors, add border...

```python
plt.figure(linewidth=5, edgecolor='black', facecolor='white')
venn2([films_i_like, films_megan_likes], set_labels=('films I like', 'films Megan likes'), 
      set_colors=('purple', 'skyblue'), alpha=0.7)
plt.title('Comprehensive Comparison of Filmography')
plt.show()
```

Adding borders around the overlapping circles may also be useful, and is done with the "venn2_circles" method

```python
venn2((10, 5, 2))
venn2_circles((10, 5, 2))
plt.show()
```

The borders can also be changed to dashed lines, etc.

```python
plt.figure(linewidth=5, edgecolor='black', facecolor='grey')
venn2((10, 5, 2), set_colors=('purple', 'skyblue'))
venn2_circles((10, 5, 2), linestyle='dashed')
plt.show()
```

The different parts of the circles, as well as the circle borders, can be singled out and their properties altered individually

```python
plt.figure(linewidth=5, edgecolor='black', facecolor='grey')
v = venn2((10, 5, 2))
c = venn2_circles((10, 5, 2))
v.get_patch_by_id('10').set_color('purple')
v.get_patch_by_id('11').set_color('orange')
v.get_patch_by_id('01').set_color('skyblue')
c[1].set_linestyle('dashed')
plt.show()
```

Also noteworthy is that the "venn2" and "venn2_circles" methods function independently

```python
plt.figure(linewidth=5, edgecolor='black', facecolor='white')
venn2(subsets=(10, 5, 2))
venn2_circles(subsets=(5, 5, 5))
plt.show()
```
